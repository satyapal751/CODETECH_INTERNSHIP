## Popularity based Recommendation System

we will reco top 50 books with heighest avg rating , and we will consider books which have 250+ user ratings.

```python
ratings.head(2)
```

```python
books.head(2)
```

```python
merged = ratings.merge(books, on='ISBN')
```

```python
x = merged.groupby(['Book-Title']).count()
```

```python
number_of_rating_df = x['Book-Rating'].reset_index()
```

```python
number_of_rating_df.rename(columns={'Book-Rating' : 'Num-Rating'}, inplace=True)
number_of_rating_df
```

```python
avg_rating_df = merged[['Book-Title', 'Book-Rating']].groupby('Book-Title').mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating' : 'Avg-Rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values(by='Avg-Rating', ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
popular_df
```

```python
popular_df.drop(['ISBN','Publisher','Image-URL-S','Image-URL-M'],axis=1, inplace=True)
```

## Collaborative based Recommendation System

```python
x = merged.groupby('User-ID').count()['Book-Rating'] > 200
padhe_like_users = x[x].index
```

```python
filterd_df = merged[merged['User-ID'].isin(padhe_like_users)]
filterd_df
```

```python
y = filterd_df.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_df = filterd_df[filterd_df['Book-Title'].isin(famous_books)]
```

```python
final_df
```

```python
pt = final_df.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recomend(name):
    # index fetch
    index = np.where(pt.index==name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])), key= lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_books:
        item = []
        temp = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-L'].values))
        data.append(item)
    return data
```

```python
data = recomend('The Notebook')
data
```

```python
for book in data:
    print(book)
```
