# Internship Task 1: Decision Tree Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('StudentsPerformance.csv')
```

```python
data.head()
```

```python
data['pass_math'] = data['math score'] >= 50
print(data['pass_math'].value_counts())
data[['math score', 'pass_math']].head()
```

```python
categorical_features = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
X = pd.get_dummies(data[categorical_features])
X.head()
```

```python
y = data['pass_math']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test data: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
plt.figure(figsize=(120, 72))
plot_tree(clf,
          feature_names=X.columns,
          class_names=['Fail', 'Pass'],
          filled=True,
          rounded=True,
          fontsize=12)
plt.title("Decision Tree Visualization for Predicting Math Pass/Fail")
plt.show()
```

# Summary

- The Decision Tree classifier achieved an accuracy of approximately **79%** on the test data.
- The model performs better in predicting students who pass math (higher recall and precision).
- Some misclassification occurs, especially predicting students who fail math (due to class imbalance).
- Features such as `test preparation course`, `parental level of education`, and `lunch` influence the predictions.
- The tree visualization helps interpret how the model makes decisions based on different categorical features.
